### (A) Descriptive Statistics:

What is the average, minimum, and maximum salary offered for data science jobs in India?

```python
import pandas as pd

# Assuming your dataset is stored in a DataFrame named 'df'
# Filter the dataset for data science job titles
data_science_jobs = df[df['job_title'].isin(['Data Scientist', 'Senior Data Scientist', 'Machine Learning Engineer', 'Data Engineer', 'Data Analyst', 'Data Architect'])]

# Calculate the average, minimum, and maximum salary offered
average_salary = data_science_jobs['avg_salary'].mean()
min_salary = data_science_jobs['min_salary'].min()
max_salary = data_science_jobs['max_salary'].max()

print("Average Salary for Data Science Jobs in India:", average_salary)
print("Minimum Salary for Data Science Jobs in India:", min_salary)
print("Maximum Salary for Data Science Jobs in India:", max_salary)
```
