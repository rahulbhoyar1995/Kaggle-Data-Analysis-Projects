```python
#Which city hosted the most matchesd


df1['city'].value_counts().idxmax()
```

```python
# How many matches were affected by the Duckworth-Lewis method?
df1['dl_applied'].value_counts()[1]
```

```python
#What was the average number of runs scored by the winning team?
df1['win_by_runs'].sum()/df1['win_by_runs'].count()
```

```python
#Which team won the most matches?
df1['winner'].value_counts().idxmax()
```

```python
# Which player has won the most "Man of the Match" awards?
df1['player_of_match'].value_counts()[0:1]
```

```python
#How many matches did each team win in a specific season?
all_seasons = list(df1['season'].unique())
for season in all_seasons:
        matches = df1[df1['season'] ==  season ]
        print(f" Season {season} : Total matches played {len(matches)}")
        print(matches['winner'].value_counts())
        print("#"*20)
```

```python
# How many times did a team win the toss and the match?
len(df1[df1['toss_winner'] == df1['winner']])
```

```python
# What is the win percentage of teams winning the toss?
len(df1[df1['toss_winner'] == df1['winner']]) / len(df1) * 100
```

```python
# How many matches were won by the team batting first?
len(df1[df1['win_by_runs'] != 0 ])
```

```python
#How many matches were won by the team batting second?
len(df1[df1['win_by_runs'] == 0 ])
```

```python
#How many matches were won by a margin of more than 50 runs?
len(df1[df1['win_by_runs'] >  50 ])
```

```python
#How many matches were won by a margin of more than 5 wickets?

len(df1[df1['win_by_wickets'] >  5 ])
```

```python
df1['city'].unique()
```

```python
#Which venue hosted the most number of matches
df1[['venue','city']].value_counts()
```

```python
#How many times did each team win the IPL championship?
winners = []
for season in sorted(list(df1['season'].unique())):
    every_season_last_match = df1[df1['season'] == season]
    winners.append(every_season_last_match.iloc[-1]['winner'])
print(pd.Series(winners).value_counts())
    
    
```

```python

#How many matches were played in each season?
for season in sorted(list(df1['season'].unique())):
    every_season = df1[df1['season'] == season]
    print(f"Season :{season} , matches_played: {len(every_season)}")
    
```

```python
#How many matches were affected by the weather (DL applied or abandoned)?

df1['dl_applied'].value_counts()[1]
```

```python


#How many times did a team win both the toss and the match by choosing to bat?
df2 = df1[df1['toss_winner'] == df1['winner']]
len(df2[df2['win_by_runs'] > 0])

```

```python
#How many times did a team win both the toss and the match by choosing to field?
df2 = df1[df1['toss_winner'] == df1['winner']]
len(df2[df2['win_by_wickets'] > 0])
```

```python


# How many times did a team successfully chase a target?

len(df1[df1['win_by_wickets'] > 0])
```

```python



#How many matches were won by the team with the highest margin of victory?
margin = 100
len(df1[df1['win_by_runs'] > margin])
```

```python


#How many times did the team batting first win in a particular season?


all_seasons = sorted(list(df1['season'].unique()))
for season in all_seasons:
    season_data = df1[df1['season'] == season]
    winner = len(season_data[season_data['win_by_runs'] > 0])
    print(f"season : {season} and number of winner : {winner}")
  
```

```python
#How many times did the team batting second win in a particular season?


all_seasons = sorted(list(df1['season'].unique()))
for season in all_seasons:
    season_data = df1[df1['season'] == season]
    winner = len(season_data[season_data['win_by_wickets'] > 0])
    print(f"season : {season} and number of winner : {winner}")
```
